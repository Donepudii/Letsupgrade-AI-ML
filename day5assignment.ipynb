# question 1
Write a Python program to find the first 20 non-even prime natural numbers.

```
a=int(input())
for i in range(a+1):
  if i>1:
    for num in range(2,i):    
      if i%num==0:
        break
    else:
      print(i)
```

# question 2
Write a Python program to implement 15 functions of string.

```
s="Hai this is b@laji."
print(s.capitalize())
print(s.casefold())
print(s.count('a'))
print(s.find('s'))
print(s.index('j'))
print(s.startswith('H'))
print(s.endswith('b'))
print(s.isalnum())
print(s.isalpha())
print(s.isspace())
print(s.lower())
print(s.upper())
print(s.find('t'))
print(s.split())
print(s.strip())
print(s.swapcase())
```

# question 3
Write a Python program to check if the given string is a Palindrome or Anagram or None of them. Display
the message accordingly to the user.


```
strr=input()
if strr==strr[::-1]:
  print("Given string is a palindrome")
else:
  print("none of them")
```

convert it finally to lower case using built-in lower(). eg: If the string is "Dr. Darshan Ingle @AIML Trainer",
then the output be "drdarshaningleaimltrainer".

```
def strr():
  a="Dr.Darshan Ingle @AIML Trainer"
  aln=""
  for i in a:
    if i.isalnum():
      aln+=i
  print(aln.lower())

strr()
```
